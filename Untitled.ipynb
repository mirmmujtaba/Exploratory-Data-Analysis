```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = pd.DataFrame({'2009' : [129.5, 86.4], '2010' : [110.4, 93.3], '2011' : [108.6, 101.7]}, index = ['People on welfare', 'People with a full time job'])
x
```

```python
plt.figure(figsize = (8, 6))
plt.title('Year 2009', fontsize = 15)
plt.bar(height = x['2009'], x = x.index)
plt.ylim(80)
plt.show()
plt.figure(figsize = (8, 6))
plt.title('Year 2010', fontsize = 15)
plt.bar(height = x['2010'], x = x.index)
plt.ylim(80)
plt.show()
plt.figure(figsize = (8, 6))
plt.title('Year 2011', fontsize = 15)
plt.bar(height = x['2011'], x = x.index)
plt.ylim(80)
plt.show()
```
