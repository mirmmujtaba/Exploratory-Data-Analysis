```python
# Plot the percentage contribution of the primary, secondary and tertiary sectors as a 
# percentage of the total GDP for all the states

GDP_list = []
for i in result1[result1['Item'] == 'Gross State Domestic Product'].transpose()[1:].values:
    for p in i:
        GDP_list.append(p)
GDP_list = pd.Series(GDP_list)

from textwrap import wrap

states = pd.Series(result1.columns)[1:]
index = np.arange(len(states))

primary = []
for i in result1[result1['Item'] == 'Primary'].transpose()[1:].values:
    for p in i:
        primary.append(p)
primary = 100*((pd.Series(primary))/(GDP_list))

secondary = []
for i in result1[result1['Item'] == 'Secondary'].transpose()[1:].values:
    for p in i:
        secondary.append(p)
secondary = 100*((pd.Series(secondary))/(GDP_list))

tertiary = []
for i in result1[result1['Item'] == 'Tertiary'].transpose()[1:].values:
    for p in i:
        tertiary.append(p)
tertiary = 100*((pd.Series(tertiary))/(GDP_list))

states = ['\n'.join(wrap(state, 10)) for state in states]

plt.figure(figsize = (27, 10))

bar_primary = plt.bar(x = index, height = primary, width = 0.5, label = 'Primary')
bar_secondary = plt.bar(x = index, height = secondary, width = 0.5, bottom = primary, label = 'Secondary')
bar_tertiary = plt.bar(x = index, height = tertiary, width = 0.5, bottom = secondary+primary, label = 'Tertiary')

plt.legend((bar_primary, bar_secondary, bar_tertiary), ('Primary', 'Secondary', 'Tertiary'))

plt.xlabel('States')
plt.ylabel('Primary, Secondary, Tertiary')

plt.xticks(index, states)
plt.autoscale()
```
